Mtodo IQR

```python
#creamos limites usando metodo de rangos inter cuartiles

#creamos la funcion  para aplicar los limites  a las columnas del dataset


#metodo de rangos inter cuartiles


for col in data_clean.columns:
    column = data_clean[col]

    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1

    umbral_inferior = Q1 - 1.5 * IQR
    umbral_superior = Q3 + 1.5 * IQR

    data_clean[col + '_outliers'] = (column < umbral_inferior) | (column > umbral_superior)

    print(f'{col =} {Q1 =} {Q3=} {IQR =} {umbral_inferior = } {umbral_superior = }')
```

```python
#quitamos los outliers y agregamos el resultante a un nuevo dataset
data_noout =  data_clean[data_clean['Total_Trans_Ct_outliers'] == False]
data_noout = data_noout[data_noout['Credit_Limit_outliers'] == False]
data_noout = data_noout[data_noout['Total_Revolving_Bal_outliers'] == False]
data_noout.reset_index(drop=True)
```

```python
#importamos librerias para clusters
from sklearn.cluster import KMeans
```

```python
#creamos funcion para #itera el modelo kmeans y almacenar los resultados en las variables definidas en el bloque anterior
inercia = []
ks = []

for k in range(1,30):
    modelo_iterado = KMeans(n_clusters = k, random_state = 10)
    modelo_iterado.fit(data_bc)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

```python
#generamos un diccionario con los valores de las inercias y ks 
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
codo
```

```python
#visualizamos el resultado del metodo de codo
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

```python
#creamos el modelo con el numero de clusters
modelo = KMeans(n_clusters=5,random_state = 10)
modelo.fit(data_bc)
```

```python
#creanos una nueva columna paara clusters
data_noout['cluster'] = modelo.labels_
data_noout
```
